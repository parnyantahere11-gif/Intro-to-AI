### 1. First Example of ML

```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
```

```python
## Generate the dataset

data = { 'Square footage (ft^2)' : [1500, 2000, 1800, 2500, 2200],
         'Number of bedrooms' : [3, 4, 3, 5, 4],
         'Number of bathrooms' : [2, 2, 1, 3, 2],
         'Garage capacity' : [2, 2, 1, 2, 2],
         'Price(USD)' : [250000, 320000, 200000, 420000, 360000] }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
## Seperate the target variable from the predictors

X = df.drop('Price(USD)', axis =1 )
y = df['Price(USD)']
```

```python
## Create a linear regression model and fit it to the data

model = LinearRegression()
model.fit(X, y)
```

```python
## Predict the price for a new house 

new_house = pd.DataFrame([[2000, 3, 2, 1]], columns = X.columns)

predicted_price = model.predict(new_house)
```

```python
print('Predicted Price is :' , predicted_price)
```



### 2. A simple code for Plotting using AI



```python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
## Data samples

df = pd.DataFrame({
    "Square footage (ft^2)": [1500, 2000, 1800, 2500, 2200],
    "Price (USD)": [250000, 320000, 280000, 420000, 360000]
})
```

```python

X = df[["Square footage (ft^2)"]]
y = df["Price (USD)"]

model = LinearRegression()
model.fit(X, y)
```

```python
## Predict a range to plot 

x_line = pd.DataFrame({"Square footage (ft^2)": range(1400, 2601, 10)})
y_line = model.predict(x_line)

plt.figure(figsize=(6,4))
plt.scatter(X, y)
plt.plot(x_line, y_line)
plt.title("Linear Regression")
plt.xlabel("Square footage (ft^2)")
plt.ylabel("Price (USD)")
plt.show()
```

```python

from textblob import TextBlob
```

```python
## Define a text for sentiment analysis

text = "I love this product! It works great and makes my life so much easier."
```

```python
## Create a Textblob object

blob = TextBlob(text)
```

```python
## Get the sentiment polarity (-1 to 1) and subjectivity (0 to 1)

polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity
```

```python
## Print the sentiment analysis results

print("Sentiment polarity:" , polarity)
print("Sentiment subjectivity:", subjectivity)

if polarity > 0:
    print(" The text is positive.")
elif polarity < 0:
    print("The text is negative.")
else:
    print("The text is neutral.")
    
```

```python

import matplotlib.pyplot as plt
import numpy as np
```

```python
def draw_heart():
    # Define the heart shape
    t = np.linspace(0, 2*np.pi, 300)
    x = 19 * np.sin(t)**3
    y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)
    
    # Plot the heart shape 
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(x, y, color='red')

     # Write the message inside the heart
    message = "Movafagh Bashid!    Data RoadMap"
    ax.text(0, 1, message, fontsize=13, ha='center', va='center')

    # Set plot properties
    ax.axis('off')
    ax.set_aspect('equal')
    plt.show()
```

```python
## Call the function to draw the heart and write the message

draw_heart()
```
